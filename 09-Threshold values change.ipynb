```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split()
from sklearn.preprocessing import StandardScaler,LabelEncoder
le = LabelEncoder()
ss = StandardScaler()
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data = pd.read_csv("C:/Users/HP/OneDrive/Desktop/DataAnalysisFiles/Datasets/breast_disease_data.csv")
data
```

```python
data["diagnosis"]=le.fit_transform(data["diagnosis"])
data.head()
```

```python
X = data.iloc[:,2:]
y = data.iloc[:,1]
```

```python
X.head()
```

```python
#Train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Standardization

```python
#Feature Scaing - standardization
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
X_train
```

```python
# LogisticRegession
log = LogisticRegression()
log.fit(X_train,y_train)
log_pred = log.predict(X_test)

accuracy_score(y_test,log_pred)
```

```python
pd.crosstab(y_test,log_pred)
```

```python
confusion_matrix(y_test,log_pred)
```

```python
log_pred
```

```python
log.predict_proba(X_test)
```

```python
log_pred[:10]
```

```python
np.round(log.predict_proba(X_test),3)[:10]
```

```python
threshold = 0.645
new_pred = np.where(log.predict_proba(X_test)[:,0]>= threshold, 0, 1  )
print(accuracy_score(y_test,new_pred))
confusion_matrix(y_test,new_pred)
```

# Getting the accuracy and confusion matrix for different threshold

```python
# Define the Threshold Function
def predict_threshold(model,X_test,threshold):
    return np.where(np.round(model.predict_proba(X_test)[:,0],3)>= threshold, 0,1)
```

```python
np.arange(0,1,0.1)
```

```python
for i in np.arange(0,1,0.1):
    y_pred = predict_threshold(log,X_test,i)
    print("Threshold value:",i)
    print("Accuracy Score",accuracy_score(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred),"\n")
    print()
```

# Bank Churners data

```python
bank = pd.read_csv("C:/Users/HP/OneDrive/Desktop/DataAnalysisFiles/Datasets/BankChurners.csv")
bank.head()
```

```python
bank.info()
```

```python
# Label Enc0ding
x = ["Attrition_Flag","Gender","Education_Level","Marital_Status","Income_Category","Card_Category"]
for i in range(len(x)):
    bank[x[i]]= le.fit_transform(bank[x[i]])
```

```python
bank.head()
```

```python
X = bank.iloc[:,2:]
y = bank.iloc[:,1]
```

```python
X.head()
```

```python
#Train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Feature scaling

```python
#Standardization
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

```python
X_train = pd.DataFrame(X_train,columns = X.columns)
X_test = pd.DataFrame(X_test,columns = X.columns)
```

# LogisticRegression

```python
log = LogisticRegression()
log.fit(X_train,y_train)
y_preds = log.predict(X_test)
accuracy_score(y_test,y_preds)

```

```python
pd.crosstab(y_test,y_preds)
```

```python
log.coef_
```

```python
a = pd.DataFrame(log.coef_.T, columns=["Coeff"])
b = pd.DataFrame(X.columns,columns = ["Col"])
c = pd.concat([b,a],axis=1)
c = c.sort_values("Coeff",ascending=False)
col = c["Col"].values[:5]
col
```

```python
X_train = X_train[col]
X_test = X_test[col]
```

```python
log.fit(X_train,y_train)
new_preds = log.predict(X_test)

accuracy_score(y_test,new_preds)
```

```python
pd.crosstab(y_test,new_preds)
```

# With Different Thresholds

```python
# Define the Threshold Function
def predict_threshold(model,X_test,threshold):
    return np.where(np.round(model.predict_proba(X_test)[:,0],3)>= threshold, 0,1)
```

```python
for i in np.arange(0.1,1,0.1):
    y_pred = predict_threshold(log,X_test,i)
    print("Threshold value:",i)
    print("Confusion Matrix: ")
    print(confusion_matrix(y_test,y_pred))
    print("Accuracy Score",accuracy_score(y_test,y_pred),"\n")
    print()
```
