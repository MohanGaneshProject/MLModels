```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(random_state=42)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.ensemble import RandomForestClassifier
clg = RandomForestClassifier(random_state=42)

#SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

#knn Imputer
from sklearn.impute import KNNImputer

import warnings
warnings.filterwarnings("ignore")
```

```python
data = {"sibsp": [1,0,1,1,0,1], "Fare": [789,723.90,4567.23,89032.9,710.34,690.12], "Age": [23,21,19,30,28,np.NaN]}
data
```

```python
X = pd.DataFrame(data)
X
```

# Iterative Imputer

```python
impute = IterativeImputer(lr)
impute.fit_transform(X)
```

```python
X.columns
```

```python
impute = IterativeImputer(reg)
pd.DataFrame(impute.fit_transform(X),columns=X.columns)
```
